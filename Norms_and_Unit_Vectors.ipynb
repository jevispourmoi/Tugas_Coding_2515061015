```python
import numpy as np
```

```python
#buat vector x dengan ukuran 3x1
x = np.array([[25], [15], [61]])
print("x:", x)
```

```python
#hitung nilai L2 norm dari vector x
L2_norm = (25**2 + 15**2 + 61**2)**(1/2)
print("L2 Norm dari vector x adalah:", L2_norm)
```

```python
#buat vector v dengan elemen (3, 4)
v = np.array([3, 4])
print("v:", v)
```

```python
#hitung nilain L2 norm dari vector v
L2_norm_v = (3**2 + 4**2)**(1/2)
print("L2 norm dari vector v adalah:", L2_norm_v)
```

```python
#tampilkan unit vector dari vector v
unit_vector_v = v / L2_norm_v
print("unit vector dari vector v adalah:", unit_vector_v)
```

```python
#hitung panjang unit vector dari vector v
panjang_unit_vector_v = np.linalg.norm(unit_vector_v)
print("panjang unit vector v adalah:", panjang_unit_vector_v)
```

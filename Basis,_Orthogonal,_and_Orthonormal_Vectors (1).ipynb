```python
import numpy as np
```

```python
#buat vector i dengan elemen (1, 0)
i = np.array([1, 0])
print("vector i =", i)
```

```python
#dan vector j dengan elemen (0, 1)
j = np.array([0, 1])
print("vector j =", j)
```

```python
#buktikan apakah vector i dan j itu orthogonal
np.dot(i, j)
```

```python
#bagaimana dengan v=(1,2) dan w=(2,-1), apakah juga orthogonal?
v = np.array([1, 2])
print("vector v:", v)
```

```python
w = np.array([2, -1])
print("vector w:", w)
```

```python
np.dot(v, w)
```

```python
#dapat dilihat bahwa vector (i, j) dan (v, w) keduanya orthogonal.
```

```python
import numpy as np
```

```python
#buat matrix X berukuran 2x3 menggunakan numpy
X = np.array([[12, 4, 7],
              [25, 15, 61]])
print("matriks X:", X)
```

```python
#tampilkan ukuran dari matrix X
X.shape
print("ukuran dari matrix X adalah:", X.shape)
```

```python
#tampilkan seluruh elemen dari kolom pertamaa dari matrix X
X[:,0]
print("seluruh elemen dari kolom pertama matrix X adalah;", X[:,0])
```

```python
#tampilkan seluruh elemen pada baris kedua dari matrix X
X[1,:]
print("semua elemen dari baris kedua matrix X adalah:", X[1,:])
```

```python
#PYTORCH
import torch
```

```python
X_pt = torch.tensor([[12, 4, 7],
              [25, 15, 61]])
print("matriks X:", X_pt)
```

```python
X_pt.shape
print("ukuran dari matriks X adalah:", X_pt.shape)
```

```python
X_pt[:,0]
print("semua elemen dari kolom pertama matrix X adalah:", X_pt[:,0])
```

```python
X_pt[1,:]
print("semua elemen dari baris kedua matrix X adalah:", X_pt[1,:])
```

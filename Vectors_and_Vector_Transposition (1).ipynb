```python
import numpy as np
```

```python
#buat vector x dengan ukuran 3x1
x = np.array([[25], [15],[61]])
x
```

```python
#hitung panjang vector x
print("panjang vector x adalah:", len(x))
```

```python
#tampilkan ukuran vector x
print("ukuran vector x adalah:", x.shape)
```

```python
#tampilkan tipe data dari vector x
print("tipe data dari vector x adalah:", type(x))
```

```python
#akses elemen pertama dari vector x
print("elemen pertama dari vector x adalah:", x[0])
```

```python
#tampikan tipe data elemen pertama dari vector x
print("tipe data dari elemen pertama vector x adalah:", type(x[0]))
```

```python
#lakukan transpose pada vector x dan simpan pada variabel x_t
x_t = x.T
print("transpose dari vector x adalah:", x.T)
```

```python
#tampilkan ukuran dari vector x_t
print("ukuran dari vektor x_t adalah:", x_t.shape)
```

```python
#vector y
y = np.array([[52, 51, 16], [12, 7, 4]])
y
```

```python
#transpose vector y
y_t = y.T
print("transpose dari vector y adalah:", y_t)
```

```python
#lakukan soal nomor 2 menggunakan pytorch dan tensorflow
import tensorflow as tf
import torch
```

```python
x_tf = tf.Variable([[25],[15],[61]], dtype=tf.int16)
x_tf
```

```python
x_pt = torch.tensor([[25],[15],[61]], dtype=torch.int16)
x_pt
```

```python
#menampilkan ukuran vector x dengan tensorflow
x_tf.shape
print("ukuran vector x adalah:", x_tf.shape)
```

```python
#menampilkan vector x dengan pytorch
x_pt.shape
print("ukuran vector x adalah:", x_pt.shape)
```

```python
#menampilkan elemen pertama dari vector x dengan pytoch
x_pt[0]
print("elemen pertama dari vector x adalah:", x_pt[0])
```

```python
#menampilkan elemen pertama dari vector x dengan tensorflow
x_tf[0]
print("elemen pertama dari vector x adalah:", x_tf[0])
```

```python
#mencari vector transpose x dengan pytorch
x_pt.T
print("transpose dari vector x adalah", x_pt.T)
```

```python
#mencari vector transpose x dengan tensorflow
x_tf_transpose = tf.transpose(x_tf)
print("transpose dari vector x adalah", x_tf_transpose)
```

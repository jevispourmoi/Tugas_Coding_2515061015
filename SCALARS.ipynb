```python
#lakukan tahapan pada soal 1 dengan menggunakan tensorflow
import tensorflow as tf
```

```python
x_tf = tf.Variable(15.0, dtype=tf.float16)
x_tf
```

```python
y_tf = tf.Variable(51.0, dtype=tf.float16)
y_tf
```

```python
tf_sum = tf.add(x_tf, y_tf)
tf_sum
```

```python
tf_float= tf.Variable(15.51, dtype=tf.float16)
tf_float
```

```python
tf_sum = tf.add(tf_float, y_tf)
tf_sum
```
